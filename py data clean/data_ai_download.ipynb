```python
import pandas as pd
import time
import random
import csv
import pyautogui
import pyperclip
```

```python
base_url= 'https://www.data.ai/search/v2?q='
bank_name = 'bank of america'
end_url = '&vertical=product'

url = base_url+bank_name+end_url
```

```python
csv_file = "bank_names_call.csv" # this is a subset of zillow_zipcodes.csv

# Read the CSV data into a Pandas DataFrame
df = pd.read_csv(csv_file)
bank_names = df.iloc[:,1].tolist()
bank_names_lower = [item.lower() for item in bank_names]
```

```python
pyautogui.moveTo(220, 440, duration=0.5)
pyautogui.doubleClick()
pyautogui.hotkey('ctrl', 'c')
clipboard_content = pyperclip.paste()
print(clipboard_content)
```

```python
for bank_name in bank_names_lower[4210:4724]:

    clipboard_content = "error"

    url = base_url+bank_name+end_url
    print(url)

    pyautogui.moveTo(250, 70, duration=0.5)
    pyautogui.click()
    pyautogui.hotkey('ctrl', 'a')
    pyautogui.press('delete')

    pyautogui.typewrite(url)
    pyautogui.press('enter')

    time.sleep(random.uniform(10, 20))
    #
    # pyautogui.moveTo(450, 340, duration=0.5)
    # pyautogui.click()
    # time.sleep(random.uniform(10, 20))
    pyautogui.moveTo(220, 440, duration=0.5)
    pyautogui.doubleClick()
    pyautogui.hotkey('ctrl', 'c')
    clipboard_content = pyperclip.paste()

    if clipboard_content != 'Showing' and clipboard_content != 'Showing ':
        print('break')
        break

    pyautogui.moveTo(900, 490, duration=0.5)
    pyautogui.click()
    pyautogui.moveTo(890, 560, duration=0.5)
    pyautogui.click()
    time.sleep(random.uniform(10, 20))

    pyautogui.hotkey('ctrl', 'w')
    time.sleep(random.uniform(10, 20))
```
